```python
import pandas as pd
```

```python
df = pd.read_csv("diabetes_complete.csv", header=0, sep=",")
```

```python
df
```

```python
df = df[df["smoking_history"] != "No Info"]
```

```python
df = df[df["gender"] != "Other"]
```

```python
dummy_table = pd.get_dummies(df, drop_first = True)
```

```python
dummy_table
```

```python
dummy_table.columns = ["age", "hypertension", "heart_disease", "bmi", "HbA1c_level", "blood_glucose_level", "diabetes", "gender", "smoking_history"]
```

```python
dummy_table = dummy_table[["gender", "age", "hypertension", "heart_disease", "bmi", "smoking_history", "HbA1c_level", "blood_glucose_level", "diabetes"]]
```

```python
df_diabetes_yes = dummy_table[dummy_table["diabetes"] == 1]
df_diabetes_no = dummy_table[dummy_table["diabetes"] == 0]
```

```python
df_final1 = df_diabetes_no.sample(n=800, random_state=1)
```

```python
df_final2 = df_diabetes_yes.sample(n=300, random_state=1)
```

```python
df_merged = df_final1.append(df_final2)
```

```python
#df_merged = df_merged[["gender", "age", "hypertension", "heart_disease", "bmi", "smoking_history", "diabetes"]]
```

```python
df_merged.index += 1 
df_merged.index.names = ['patient_id']
df_merged = df_merged.astype({'age':'int'})
```

```python
df_merged.to_csv("diabetes_double.csv", index=True, header=True, sep=",")
```
